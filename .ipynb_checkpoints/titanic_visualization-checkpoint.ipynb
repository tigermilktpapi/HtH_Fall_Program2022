```python
import pandas
import seaborn
import matplotlib.pyplot as plt

''' 
The best way to complete this activity is to copy the code in this file into a jupyter notebook file you create. This will also you to see your visualizations easier. Once you have finished creating your visuals copy your code back into this file and submit it to Replit. Don't forget to also upload your new code to Github! 
'''

# Call in our titanic csv
# Note use a .. first if this file is in another folder but because it's on the same level as the other folders the .. is not needed 
titanic_data = pandas.read_csv("Pandas-Data-Manipulation/titanic.csv")
titanic_data
```

```python
# Step 1: Create a heatmap showing the correlation between the following features: survival, pclass, sex, age, and fare 
titanic_features = titanic_data[["Survived", "Pclass", "Age", "Fare"]]


corr_heatmap = seaborn.heatmap(titanic_features.corr(), annot = True, cmap = "mako")
corr_heatmap
```

```python
# Step 2: Create a bar graph that illustrates how many passengers fell into each of these age ranges: 0 - 16, 17 - 25, 26 -40, 41 - 59, 60 or older
ages = ["16 or younger", "17 to 25", "26 to 40", "41 to 60", "60+"]
age_groups = {
    
    "groups" : ages,
    "count" : [
        len(titanic_data[titanic_data["Age"] <= 16.0]),
        len(titanic_data[(titanic_data["Age"] >= 17.0) & (titanic_data["Age"] <= 25.0)]),
        len(titanic_data[(titanic_data["Age"] >= 26.0) & (titanic_data["Age"] <= 40.0)]),
        len(titanic_data[(titanic_data["Age"] >= 41.0) & (titanic_data["Age"] <= 59.0)]),
        len(titanic_data[(titanic_data["Age"] >= 60.0)])],
}
vis_df = pandas.DataFrame(age_groups)
seaborn.barplot(data = vis_df, x = "groups", y = "count", palette = "rocket")
plt.show()
```
