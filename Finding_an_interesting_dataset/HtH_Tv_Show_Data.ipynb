```python
import pandas

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

tvshow = pandas.read_csv("tv_shows.csv")
tvshow
```

```python
netflix = tvshow[tvshow["Netflix"] == 1]
nshows = len(netflix)

hulu = tvshow[tvshow["Hulu"] == 1]
hshows = len(hulu)

primevideo = tvshow[tvshow["Prime Video"] == 1]
pshows = len(primevideo)

disney = tvshow[tvshow["Disney+"] == 1]
dshows = len(disney)

print("Netflix show count", nshows,"\nHulu show count", hshows, "\nPrime Video show count", pshows, "\nDisney+ show count", dshows)
```

# Finding which streaming platform had the most tv shows across Netflix, Hulu, Prime Video, and Disney+. My findings show that Netflix had the most tv shows total at 1971 with Prime Video at a close second with 1831 shows on their platform. That being said Disney+ had the least amount of shows at 351.
